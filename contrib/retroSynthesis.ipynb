```python
from aizynthfinder.aizynthfinder import AiZynthFinder, AiZynthExpander
```

```python
filename = "./../data/config.yml"
finder = AiZynthFinder(configfile=filename)
```

```python
finder.stock.select("zinc")
finder.expansion_policy.select("uspto")
finder.filter_policy.select("uspto")
```

```python
moroidin = 'CC(C)CC1C(=O)NC(C(=O)NC2CC3=C(NC4=C3C=CC(=C4)C(C(C(=O)N1)NC(=O)C5CCC(=O)N5)C(C)C)N6C=C(CC(NC(=O)CNC(=O)C(NC2=O)CCC[NH+]=C(N)N)C(=O)[O-])N=C6)C(C)C'
```

```python
finder.target_smiles = moroidin
finder.tree_search()
```

```python
finder.build_routes()
stats = finder.extract_statistics()
```

```python
expander = AiZynthExpander(configfile=filename)
expander.expansion_policy.select("uspto")
expander.filter_policy.select("uspto")
reactions = expander.do_expansion(moroidin)
```

```python
reactants_smiles = []
for reaction_tuple in reactions:
    reactants_smiles.append([mol.smiles for mol in reaction_tuple[0].reactants])
```

```python
import pandas as pd
metadata = []
for reaction_tuple in reactions:
    for reaction in reaction_tuple:
        metadata.append(reaction.metadata)
df = pd.DataFrame(metadata)
```
